```python
import pandas as pd

import os
os.listdir('./fraud_data')

df = pd.read_csv('./fraud_data/Base.csv')
df.shape, df.columns

df.info()
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder

cat_cols = df.select_dtypes(include=['object']).columns
print("Categorical columns:", cat_cols.tolist())

le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in ['payment_type', 'employment_status', 'housing_status', 'source', 'device_os']:
    df[col] = le.fit_transform(df[col])
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('fraud_bool', axis=1)
y = df['fraud_bool']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

print("Size of train set:", X_train.shape)
print("Size of test set:", X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix


models = {
    "Logistic Regression": LogisticRegression(class_weight='balanced', max_iter=3000, random_state=42),
    "Random Forest": RandomForestClassifier(class_weight='balanced', n_estimators=100, random_state=42),
    "XGBoost": XGBClassifier(scale_pos_weight=(y_train == 0).sum() / (y_train == 1).sum(),
                              use_label_encoder=False, eval_metric='logloss'),
    "MLP": MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=100, random_state=42)
}

results = {}

for name, model in models.items():
    print(f"\nTraining {name}...")
    if name in ["Logistic Regression", "MLP"]:
        model.fit(X_train_scaled, y_train)
        y_pred = model.predict(X_test_scaled)
        y_prob = model.predict_proba(X_test_scaled)[:, 1]
    else:
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:, 1]

    print(classification_report(y_test, y_pred, digits=4))
    print("ROC AUC Score:", roc_auc_score(y_test, y_prob))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
import numpy as np
import shap
import matplotlib.pyplot as plt

# ç¡®ä¿ä½ å·²ç»æœ‰è¿™äº›å˜é‡
# y_test, y_pred_xgb, y_prob_xgb, shap_values, X_test, models["XGBoost"]

# é€‰æ‹©å‰3ä¸ª True Positives å’Œ False Positives
tp_indices = np.where((y_test.values == 1) & (y_pred_xgb == 1))[0][:3]
fp_indices = np.where((y_test.values == 0) & (y_pred_xgb == 1))[0][:3]

# å®šä¹‰å¯è§†åŒ– + è¾“å‡ºå‡½æ•°
def explain_local(idx, title):
    print(f"\n===== {title} | Sample Index: {idx} =====")
    print(f"âœ… Predicted Prob (Fraud): {y_prob_xgb[idx]:.4f}")
    print(f"âœ… True Label: {y_test.values[idx]}")

    # å±•ç¤º Waterfall å›¾ï¼ˆå±•ç¤ºå‰ 5 ä¸ªé‡è¦ç‰¹å¾ï¼‰
    shap.plots.waterfall(shap_values[idx], max_display=5)

    # è¾“å‡º Top-5 ç‰¹å¾åŠ SHAP å€¼
    shap_dict = dict(zip(X_test.columns, shap_values[idx].values))
    top5 = sorted(shap_dict.items(), key=lambda x: abs(x[1]), reverse=True)[:5]

    print("\nTop 5 SHAP Features:")
    for feat, val in top5:
        direction = "â†‘ Increase Fraud Risk" if val > 0 else "â†“ Decrease Fraud Risk"
        print(f"  {feat:30s}: {val:+.3f} ({direction})")

# å¯¹ TP æ ·æœ¬è§£é‡Š
for i, idx in enumerate(tp_indices):
    explain_local(idx, f"TP Sample #{i+1}")

# å¯¹ FP æ ·æœ¬è§£é‡Š
for i, idx in enumerate(fp_indices):
    explain_local(idx, f"FP Sample #{i+1}")
```

```python
import optuna
from sklearn.metrics import f1_score
from xgboost import XGBClassifier

def objective(trial):

    param = {
        "max_depth": trial.suggest_int("max_depth", 3, 10),
        "min_child_weight": trial.suggest_int("min_child_weight", 1, 10),
        "eta": trial.suggest_float("eta", 0.01, 0.3, log=True),
        "scale_pos_weight": (y_train == 0).sum() / (y_train == 1).sum(),
        "objective": "binary:logistic",
        "eval_metric": "logloss",
        "use_label_encoder": False,
        "n_jobs": -1,
        "random_state": 42
    }

    model = XGBClassifier(**param)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return f1_score(y_test, y_pred)
```

```python
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=30)
```

```python
def evaluate_fairness_across_groups(y_true, y_pred, X_test, group_vars):
    records = []
    for group in group_vars:
        g = X_test[group]
        try:
            dp = demographic_parity_difference(y_true, y_pred, sensitive_features=g)
            eo = equalized_odds_difference(y_true, y_pred, sensitive_features=g)
            eop = equal_opportunity_difference(y_true, y_pred, sensitive_features=g)
            records.append({
                'Sensitive Feature': group,
                'DP Diff': round(dp, 3),
                'EO Diff': round(eo, 3),
                'EOp Diff': round(eop, 3)
            })
        except Exception as e:
            print(f"Warning for {group}: {e}")
    return pd.DataFrame(records)

group_candidates = ['employment_status', 'housing_status', 'device_os', 'customer_age']  # å¯è¡¥å……
fairness_summary = evaluate_fairness_across_groups(y_test, y_pred_xgb, X_test, group_candidates)
print("== å„ç¾¤ä½“å˜é‡çš„å…¬å¹³æ€§å·®è·å¯¹æ¯” ==")
print(fairness_summary)
fairness_summary.set_index('Sensitive Feature')[['DP Diff', 'EO Diff', 'EOp Diff']].plot(kind='barh', figsize=(8, 5), title='Fairness Metric Differences by Sensitive Feature')
plt.tight_layout(); plt.show()
```

```python
# === åœ¨ä»£ç ç‰‡æ®µ 11 ä¹‹å‰æ’å…¥ï¼Œåˆ†æžå¤šä¸ªç¾¤ä½“å˜é‡çš„å…¬å¹³æ€§ ===
import pandas as pd
import matplotlib.pyplot as plt
from fairlearn.metrics import demographic_parity_difference, equalized_odds_difference, equal_opportunity_difference

def evaluate_fairness_across_groups(y_true, y_pred, X_test, group_vars):
    records = []
    for group in group_vars:
        g = X_test[group]
        try:
            dp = demographic_parity_difference(y_true, y_pred, sensitive_features=g)
            eo = equalized_odds_difference(y_true, y_pred, sensitive_features=g)
            eop = equal_opportunity_difference(y_true, y_pred, sensitive_features=g)
            records.append({
                'Sensitive Feature': group,
                'DP Diff': round(dp, 3),
                'EO Diff': round(eo, 3),
                'EOp Diff': round(eop, 3)
            })
        except Exception as e:
            print(f"Warning for {group}: {e}")
            continue
    return pd.DataFrame(records)

# ðŸ” æ›¿æ¢ä¸ºä½ æ„Ÿå…´è¶£çš„ç¾¤ä½“å˜é‡åˆ—è¡¨
group_candidates = ['employment_status', 'housing_status', 'device_os', 'customer_age']

fairness_summary = evaluate_fairness_across_groups(y_test, y_pred_xgb, X_test, group_candidates)
print("== å„ç¾¤ä½“å˜é‡çš„å…¬å¹³æ€§å·®è·å¯¹æ¯” ==")
print(fairness_summary)

# ï¼ˆå¯é€‰ï¼‰å¯è§†åŒ–
fairness_summary.set_index('Sensitive Feature')[['DP Diff', 'EO Diff', 'EOp Diff']].plot(kind='barh', figsize=(8, 5), title='Fairness Metric Differences by Sensitive Feature')
plt.tight_layout()
plt.show()

```
